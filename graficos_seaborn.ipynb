```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('dormir.csv', sep=';')

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Sleep Duration', fill=True)

plt.title('Gráfico de Densidad de Duración del Sueño')
plt.xlabel('Duración del Sueño (horas)')
plt.ylabel('Densidad')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('dormir.csv', sep=';')

plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Daily Steps'])

plt.title('Caja de Bigotes de Pasos Diarios')
plt.ylabel('Pasos Diarios')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('dormir.csv', sep=';')

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Quality of Sleep', order=df['Quality of Sleep'].value_counts().index, palette='viridis')

plt.title('Conteo de Calidad del Sueño')
plt.xlabel('Calidad del Sueño')
plt.ylabel('Conteo')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('dormir.csv', sep=';')

df_numeric = df.select_dtypes(include=np.number).drop(columns=['Person ID'])

df_numeric = df_numeric.fillna(0)

plt.figure(figsize=(10, 8))
sns.clustermap(df_numeric.corr(), annot=True, cmap='viridis', fmt=".2f")

plt.title('Mapa de Calor con Clustering', pad=20)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('dormir.csv', sep=';')

disorder_categories = ['Sleep Apnea', 'None', 'Insomnia']
df_filtered = df[df['Sleep Disorder'].isin(disorder_categories)].copy()

plt.figure(figsize=(12, 8))
sns.violinplot(data=df_filtered, x='Sleep Disorder', y='Age', hue='Occupation', split=False)

plt.title('Gráfico de Violín de la Edad por Trastorno del Sueño y Ocupación')
plt.xlabel('Trastorno del Sueño')
plt.ylabel('Edad')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
